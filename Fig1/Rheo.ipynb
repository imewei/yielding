```python
def read_parameters(fname):
    if fname.endswith('.txt'):
        paras = np.loadtxt(fname)
    elif fname.endswith('.csv'):
        paras = np.loadtxt(fname, delimiter=',')
    else:
        paras = None
    return paras

def cumulative_sum(array):
    result = []
    total = 0
    for num in array:
        total += num
        result.append(total)
    return result
```
